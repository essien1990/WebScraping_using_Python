# Web Scraping using Python BeautifulSoup

```python
# A browser's dev tools to learn the structure of any website when scraping
# Libraries like beautifulsoup and requests and Pandas will be used
# Task: Pull or scrape this data or some part of the data and put it into an excel Sheet by Using Pandas Dataframe
        # Get this week's forecast to know today's weather, Friday's weather, Sunday's weather,etc.
        # Descriptions of the Weather
        # Low or Higher Temperature of the weather for each day
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# Request the URL
# https://www.crummy.com/software/BeautifulSoup/bs4/doc/#installing-a-parser
page = requests.get('https://forecast.weather.gov/MapClick.php?lat=34.0536&lon=-118.2454')
#create a soup object of the markup or webpage using beautifulsoup
soup = BeautifulSoup(page.content, 'lxml')
# Display the source code of the page
soup
```

```python
# Find all the links(anchor tags)
soup.find_all('a')
```

```python
# Find the mother div that contains all the forecasts
weather = soup.find(id='seven-day-forecast-container')
```

```python
# print the weather forecasts, using .text() or .prettify() inside that div
weather
```

```python
# Access each item under from the list(<li></li>) the div containing the id 'seven-day-forecast-container'
# Take note unordered list(<ul></ul>). 
# The items can be either access by using the li or getting the class 'forecast-tombstone'
# In beautifulSoup, the class cannot be used because is special in Python so class_ must be used
```

```python
weather.find_all(class_='forecast-tombstone')
```

```python
forecast_items = weather.find_all(class_='tombstone-container')
```

```python
forecast_items
```

```python
# Getting the first forcast item
forecast_items[0]
```

```python
# Get all the period-name, short-desc and temp and put all in separate columns
# Get the text of the period name class
forecast_items[0].find(class_='period-name').get_text(' ')
```

```python
# Get the text of the short description class
forecast_items[0].find(class_='short-desc').get_text()
```

```python
# Get the text of the temperature class
forecast_items[0].find(class_='temp').get_text()
```

```python
# Collect all forecast items in the list by running a for loop(list comprehension)
# get for period name
period_names = [item.find(class_='period-name').get_text(' ') for item in forecast_items]
period_names
```

```python
# get for short descriptions
short_descriptions = [item.find(class_='short-desc').get_text(' ') for item in forecast_items]
short_descriptions
```

```python
# get for temperature
temperature = [item.find(class_='temp').get_text(' ') for item in forecast_items]
temperature
```

```python
# Create a dataframe with new column names
weather_data = pd.DataFrame(
{
    'period':period_names,
    'short_descriptions':short_descriptions,
    'temperature':temperature
})
```

```python
# print the new dataframe or data
weather_data
```

```python
# export data to CSV
weather_data_csv = weather_data.to_csv('weather-forecast.csv', index=False,header=True)
```

```python
# export data to Json
weather_data_json = weather_data.to_json('weather-forecast.json')
```
