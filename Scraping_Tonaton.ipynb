```python
# Print all text in the paragraph tag
# -- list dictionaries
# paras = tag.text for tag in paragraphs] 
for tag in soup.find_all('p'):
    print(tag.text)
```

# Scraping the House Listings on Tonaton

```python
# get the requests of the URL
results = requests.get('https://tonaton.com/en/ads/ghana/houses-for-sale')
```

```python
# Check if the status code is successful
results.status_code
```

```python
# Soupify the Markup Content
soup = BeautifulSoup(results.content, 'lxml')
```

```python
# view the markup or content
soup
```

```python
listings = soup.find_all('li', {'class': 'top-ads-container--1Jeoq'})
```

```python
listings[0].text.split('|')
```

```python
for house in listings:
    print(house.text)
```

```python
# Slipt the below data with split(',') but still looks way off
    # property description: Western, Houses For Sale,
    # Beds: 6
    # Baths: 5
    # Price: 4000,000
listings[0].text.split(',')
```

```python
# Import regex 
import re
```

```python
# Ocean front, 70 m of lagoon frontage, close to Old Protea HotelBeds: 6, Baths: 5Western, Houses For SaleGH₵ 4,000,000
# Make it look nice by including whitespaces with a comma ',' between Hotel and Beds by separating them using replace()
listings[0].text.replace('Beds: ', ', Beds: ')
```

```python
# replace Baths, Here the Regex comes in
    #/d -- regualr expressioin for number
    # + -- Denotes more that pattern
    #Filter numbers but a multiple of them
re.sub(r'Baths: (\d+)', r'Baths: \1, ', listings[0].text)
```

```python
# The regex is used to replace all occurences of baths and the number associated with it
# with same pattern but with a comma and space appended to them
re.sub(r'Beds: (\d+)', r' Beds: \1, ', listings[0].text)
```

```python
# Clean for beds, baths using a function
def clean(text):
    text = re.sub(r'Baths: (\d+)', r'Baths: \1, ', text)
    text = re.sub(r'Beds: (\d+)', r' Beds: \1, ', text)
    text = text.split(',')
    return text
```

```python
for house in listings:
    print(clean(house.text))
```
